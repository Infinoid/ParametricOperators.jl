```julia
using Base64
using LinearAlgebra
using Random
using Revise
using ParametricOperators
using Printf
using PyCall
using Zygote

@pyimport matplotlib.animation as anim
using PyPlot
```

```julia
# Global parameters
N = 2     # Number of exponentials
Î´ = 0.1   # Spacing between observations
t = 0:Î´:1 # Observation points
t = collect(t)

# Model given by sum of exponentials
f(c, Î±; x=t) = sum(map(i -> c[i].*exp.(Î±[i].*x), 1:N))

# Create ground truth f(x) and noisy observations y
Random.seed!(123)
Î±t = 10*randn(N)
ct = rand(N)
Ï  = 0.05
y  = f(ct, Î±t) + Ï.*randn(length(t))

@show Î±t ct

figure()
scatter(t, y, marker="x", color="red")
```

```julia
# Setup parametric function (linear in first argument, nonlinear in second argument)
Î¦ = ParLinearFunction(f, N; P=Parametric)
```

```julia
# Inner solve for linear parameters using Nesterov's method
function cÌ(A, c=rand(N); Î·=0.01, max_iter=10000, tol=1e-5)
    c_prev = zero(c)
    for i = 1:max_iter
        
        # Nesterov momentum
        p = (i-1)/(i+2).*(c .- c_prev)
        
        # Compute the gradient w.r.t c plus momentum
        # â/âc ||y - A*(c+p)||â
        g = gradient(x -> norm(y - A*(x.+p)), c)[1]
        if (norm(g)) â¤ tol
            break
        end
        
        # Nesterov update câââ = câ + pâ - Î·*âf(câ+pâ)
        c .+= p
        c .-= Î·.*g
        c_prev .= c
    end
    return c
end
```

```julia
# Outer solve for nonlinear parameters
Î± = randn(N)
Î· = 0.01
max_iter = 2000
tol = 1e-3

# Store iterates
cs = []
Î±s = []
rs = []

for i = 1:max_iter
    
    # Solve the convex problem for c at the parameters given by Î±
    # min ||Î¦(Î±)*c - y||â w.r.t c
    c = cÌ(Î¦(Î±))
        
    # Compute the jacobian w.r.t Î± on the residual r = y - Î¦(Î±)*c
    J = jacobian(a -> y - Î¦(a)*c, Î±)[1]
    r = y - Î¦(Î±)*c
    
    # Step using Jáµr
    g = J'*r
    Î± .-= g
    
    push!(cs, c)
    push!(Î±s, copy(Î±))
    push!(rs, r)
    
    # Display iterates
    if mod1(i, 100) == 1
        iterate = lpad(i,4,"0")
        println("i = $iterate, norm(y - Î¦(a)*c(Î±)) = $(norm(r))")
    end
end
```

```julia
tÌ = 0:0.01:1
chat = cs[end]
Î±hat = Î±s[end]
yÌ = f(chat, Î±hat; x=tÌ)

figure(figsize=(6, 6))
scatter(t, y, marker="x", color="red", label=L"\mathbf{y}_{obs}")
plot(tÌ, yÌ, label=L"\hat{\mathbf{y}}")
xlabel(L"t")
ylabel(L"y")
title(latexstring("Solution ", L"\mathbf{\Phi}(\mathbf{\alpha})\mathbf{c} = \sum_{i=1}^N c_i e^{\alpha_i t}"))
legend()

figure(figsize=(14, 4))

step = 100
res = collect(map(r -> norm(r), rs))
subplot(131)
plot(collect(1:length(res)), res)
scatter(1:step:length(res), res[1:step:end], marker="o", s = 10)
yscale("log")
xlabel(latexstring("Iteration ", L"k"))
ylabel(L"||\mathbf{y} - \Phi(\mathbf{\alpha})\mathbf{c}||_2^2")
title("Convergence of residual")

cs_out = hcat(cs...)
subplot(132)
plot(cs_out[1,:], cs_out[2,:], label=L"\mathbf{c}_k")
scatter(cs_out[1,1:step:end-1], cs_out[2,1:step:end-1], marker="o", s = 10)
scatter(cs_out[1,end], cs_out[2,end], marker="x", s = 20, color="red", label=L"\mathbf{\hat{c}}")
xlabel(L"c_1")
ylabel(L"c_2")
title(latexstring("Convergence of ", L"\mathbf{c}"))
legend()

Î±s_out = hcat(Î±s...)
subplot(133)
plot(Î±s_out[1,:], Î±s_out[2,:], label=L"\mathbf{\alpha}_k")
scatter(Î±s_out[1,1:step:end-1], Î±s_out[2,1:step:end-1], marker="o", s = 10)
scatter(Î±s_out[1,end], Î±s_out[2,end], marker="x", s = 20, color="red", label=L"\mathbf{\hat{\alpha}}")
xlabel(L"\alpha_1")
ylabel(L"\alpha_2")
title(latexstring("Convergence of ", L"\mathbf{\alpha}"))
legend()

tight_layout()
```

```julia
# https://genkuroki.github.io/documents/Jupyter/20170624%20Examples%20of%20animations%20in%20Julia%20by%20PyPlot%20and%20matplotlib.animation.html#Example-3
function showanim(filename)
    base64_video = base64encode(open(filename))
    display("text/html", """<video controls src="data:video/x-m4v;base64,$base64_video">""")
end

fig = figure(figsize=(6,6))

# i=0,1,...,frames-1
function animate(i)
    clf()
    tÌ = 0:0.01:1
    chat = cs[i+1]
    Î±hat = Î±s[i+1]
    yÌ = f(chat, Î±hat; x=tÌ)
    scatter(t, y, marker="x", color="red", label=L"\mathbf{y}_{obs}")
    xlim((-0.1, 1.1))
    ylim((-0.25, 1.25))
    xlabel(L"t")
    ylabel(L"y")
    title(latexstring("Solution ", L"\mathbf{\Phi}(\mathbf{\alpha})\mathbf{c} = \sum_{i=1}^N c_i e^{\alpha_i t}", " at iteration $(i+1)"))
    plot(tÌ, yÌ, label=L"\hat{\mathbf{y}}")
end

nframes = length(rs)
frametol = 1e-4
for i = 1:length(rs)-1
    if norm(rs[i+1].-rs[i]) â¤ frametol
        nframes = i
        break
    end
end

# Warning: takes a long time!
withfig(fig) do
    myanim = anim.FuncAnimation(fig, animate, frames=nframes, interval=25, blit=true)
    myanim[:save]("convergence.mp4", bitrate=-1, extra_args=["-vcodec", "libx264", "-pix_fmt", "yuv420p"])
end

showanim("convergence.mp4")
```
